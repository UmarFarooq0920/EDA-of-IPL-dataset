```python
def avrge(batsman):
    runs = df_deliveries.loc[df_deliveries['batsman']==batsman,:]['batsman_runs'].sum()
    outs = df_deliveries.loc[df_deliveries['batsman']==batsman,:]['outs'].sum()
    balls = len(df_deliveries.loc[df_deliveries['batsman']==batsman,:].index)
    if outs != 0:
        avrgee = runs/outs
    else:
        avrgee = runs
    strike_rate = (100*runs)/balls
    return batsman,runs,outs,balls,avrgee,strike_rate
```

```python
runs = df_deliveries.loc[df_deliveries['batsman']=='J Bairstow',:]['batsman_runs'].sum()
outs = df_deliveries.loc[df_deliveries['batsman']=='J Bairstow',:]['outs'].sum()
avg = runs/outs
avg
```

```python
batsman_data = pd.DataFrame()
```

```python
batsman_data
```

```python
batsman_name = []
average_list = []
runs_list = []
outs_list =[]
strike_rate = []
balls_list= []
```

```python
for batsman in df_deliveries['batsman'].unique():
    player,runs,outs,balls,average,strikerate = avrge(batsman)
    batsman_name.append(player)
    average_list.append(average)
    runs_list.append(runs)
    outs_list.append(outs)
    strike_rate.append(strikerate)
    balls_list.append(balls)
```
